```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

def safe_filename(name):
    name = re.sub(r"[^\w\-. ]+", "_", name)
    return name.replace(" ", "_")

def analyze_tokens_with_keys_pie(
    df,
    col,
    key_col="My_Key",
    fig_dir="figs",
    unique_per_paper=True,
    filename=None,
    title=None,
    min_pct=2      # hide very small slices (optional)
):
    os.makedirs(fig_dir, exist_ok=True)

    tmp = df[[key_col, col]].dropna(subset=[col]).copy()
    tmp[col] = tmp[col].astype(str)

    # split ONLY by newline
    tokens_df = (
        tmp
        .assign(Token=tmp[col].str.split("\n"))
        .explode("Token")
    )

    tokens_df["Token"] = tokens_df["Token"].str.strip()
    tokens_df = tokens_df[tokens_df["Token"] != ""]

    # remove duplicates within same paper
    if unique_per_paper:
        tokens_df = tokens_df.drop_duplicates(subset=[key_col, "Token"])

    # Token → [My_Key]
    mapping = (
        tokens_df
        .groupby("Token")[key_col]
        .apply(list)
        .sort_index()
    )

    # Frequency
    freq = tokens_df["Token"].value_counts()

    if filename is None:
        filename = safe_filename(col) + "_pie.png"
    if title is None:
        title = f"Distribution of {col}"

    # ---- PIE PLOT ----
    plt.figure(figsize=(9, 9))

    def autopct(pct):
        total = freq.sum()
        count = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n(n={count})" if pct >= min_pct else ""

    freq.plot(
        kind="pie",
        autopct=autopct,
        startangle=90,
        counterclock=False
    )

    plt.title(title)
    plt.ylabel("")  # remove default ylabel
    plt.tight_layout()

    outpath = os.path.join(fig_dir, filename)
    plt.savefig(outpath, dpi=300, bbox_inches="tight")
    plt.show()

    print("==== Token → [My_Key] ====")
    print(mapping)
    print("\n==== Frequency ====")
    print(freq)
    print(f"\n{filename} saved ✅")

    return tokens_df, mapping, freq

results_pie = {}

cols_to_run = [
    # "RQ1a_Encoding_Type",
    # "RQ1b_Learnable_encoding_parameters",
    # "RQ1c_Training_mechanism",
    "RQ1d_Hybrid_Integration_Type",
    # "RQ2a_spike_wise_efficiency",
    # "RQ2b_Resource_efficiency",
    # "RQ2c_Training_Behavior",
    # "RQ2d_Model_Behavior",
    "RQ3a_Dataset_modality",
    # "RQ3b_Input_Representation",
    # "RQ3c_Datasets_Used",
    # "RQ3d_Task_or_Domain",
    # "RQ3e_Evaluation_Metrics",
    # "RQ4a_Limitation_Category",
    # "RQ4b_Reported_Limitations",
    # "RQ4c_Research_Gaps",
    # "hardware_or_simulator",
    "paper_type",
    # Architecture_Type",
    # "code_availability",
    # "training_config_short",
]

for c in cols_to_run:
    tokens_df, mapping, freq = analyze_tokens_with_keys_pie(
        df,
        col=c,
        fig_dir=fig_dir,
        filename=f"{safe_filename(c)}_pie.png",
        title=f"Distribution of {c}"
    )
    results_pie[c] = {
        "mapping": mapping,
        "freq": freq
    }
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# =========================
# Columns
# =========================
key_col   = "My_Key"
enc_col   = "RQ1a_Encoding_Type"
spike_col = "RQ2a_spike_wise_efficiency"
res_col   = "RQ2b_Resource_efficiency"

# folder
fig_dir = "figs"
os.makedirs(fig_dir, exist_ok=True)

# =========================
# Load + Expand (paper-level)
# =========================
data = df[[key_col, enc_col, spike_col, res_col]].dropna(subset=[enc_col]).copy()

expanded = (
    data
    .assign(Encoding=data[enc_col].astype("string").str.split("\n"))
    .explode("Encoding")
)

expanded["Encoding"] = expanded["Encoding"].astype("string").str.strip()
expanded[key_col] = expanded[key_col].astype("string").str.strip()

# drop empty keys/encodings
expanded = expanded.dropna(subset=[key_col, "Encoding"])
expanded = expanded[(expanded[key_col] != "") & (expanded["Encoding"] != "")]

# =========================
# Clean spike/resource text -> true NA
# =========================
bad = {
    "<na>", "na", "nan", "none", "null",
    "n/a", "not applicable", "not reported",
    "no", "no improvement", "none reported"
}

for c in [spike_col, res_col]:
    expanded[c] = expanded[c].astype("string").str.strip()
    expanded.loc[expanded[c] == "", c] = pd.NA
    expanded.loc[expanded[c].str.lower().isin(bad), c] = pd.NA

# ✅ prevent inflated counting:
# one paper counted once per encoding
expanded = expanded.drop_duplicates(subset=[key_col, "Encoding"])

# =========================
# Flags (reported? yes/no)
# =========================
expanded["Spike_Improvement"] = expanded[spike_col].notna().astype(int)
expanded["Resource_Improvement"] = expanded[res_col].notna().astype(int)
expanded["Both_Improvements"] = (
    (expanded["Spike_Improvement"] == 1) &
    (expanded["Resource_Improvement"] == 1)
).astype(int)
expanded["Spike_Only"] = (
    (expanded["Spike_Improvement"] == 1) &
    (expanded["Resource_Improvement"] == 0)
).astype(int)
expanded["Resource_Only"] = (
    (expanded["Spike_Improvement"] == 0) &
    (expanded["Resource_Improvement"] == 1)
).astype(int)

# =========================
# Pivots (counts of papers)
# =========================
pivot_spike = pd.pivot_table(
    expanded, index="Encoding", values="Spike_Improvement", aggfunc="sum"
)
pivot_resource = pd.pivot_table(
    expanded, index="Encoding", values="Resource_Improvement", aggfunc="sum"
)

# common order
order = (
    pivot_spike.join(pivot_resource, how="outer")
    .fillna(0)
    .sum(axis=1)
    .sort_values(ascending=False)
    .index
)
pivot_spike = pivot_spike.loc[order]
pivot_resource = pivot_resource.loc[order]

# =========================
# My_Key mappings
# =========================
spike_keys = (
    expanded.loc[expanded["Spike_Improvement"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
resource_keys = (
    expanded.loc[expanded["Resource_Improvement"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
both_keys = (
    expanded.loc[expanded["Both_Improvements"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
spike_only_keys = (
    expanded.loc[expanded["Spike_Only"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
resource_only_keys = (
    expanded.loc[expanded["Resource_Only"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)

# =========================
# Sanity checks (optional but useful)
# =========================
print("Spike=1 count:", int(expanded["Spike_Improvement"].sum()))
print("Resource=1 count:", int(expanded["Resource_Improvement"].sum()))
print("Both=1 count:", int(expanded["Both_Improvements"].sum()))
print("Spike-only=1 count:", int(expanded["Spike_Only"].sum()))
print("Resource-only=1 count:", int(expanded["Resource_Only"].sum()))

# =========================
# Plot side-by-side
# =========================
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

pivot_spike.plot(kind="bar", ax=axes[0], legend=False)
axes[0].set_title("Spike-wise Efficiency Improvements (count of papers)")
axes[0].set_xlabel("Encoding Type")
axes[0].set_ylabel("Number of Papers")
axes[0].tick_params(axis="x", rotation=45)
axes[0].yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
axes[0].yaxis.grid(True, linestyle="--", alpha=0.6)

pivot_resource.plot(kind="bar", ax=axes[1], legend=False)
axes[1].set_title("Resource Efficiency Improvements (count of papers)")
axes[1].set_xlabel("Encoding Type")
axes[1].tick_params(axis="x", rotation=45)
axes[1].yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
axes[1].yaxis.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()

out_path = os.path.join(fig_dir, "Resource_Spike_wise_Efficiency_Improvements.png")
plt.savefig(out_path, dpi=300, bbox_inches="tight")
print(f"{out_path} saved ✅")
plt.show()

# =========================
# Print My_Key for discussion
# =========================
print("\n==== Encoding → My_Key (Spike-wise improvements) ====")
print(spike_keys)

print("\n==== Encoding → My_Key (Resource improvements) ====")
print(resource_keys)

print("\n==== Encoding → My_Key (Both improvements) ====")
print(both_keys)

print("\n==== Encoding → My_Key (Spike-only) ====")
print(spike_only_keys)

print("\n==== Encoding → My_Key (Resource-only) ====")
print(resource_only_keys)
```

```python
import pandas as pd
import os

key_col   = "My_Key"
enc_col   = "RQ1a_Encoding_Type"
spike_col = "RQ2a_spike_wise_efficiency"

data_spike = df[[key_col, enc_col, spike_col]].dropna(subset=[enc_col, spike_col]).copy()

expanded_spike = (
    data_spike
    .assign(
        Encoding=data_spike[enc_col].astype("string").str.split("\n"),
        Improvement=data_spike[spike_col].astype("string").str.split("\n")
    )
    .explode("Encoding")
    .explode("Improvement")
)

expanded_spike["Encoding"] = expanded_spike["Encoding"].astype("string").str.strip()
expanded_spike["Improvement"] = expanded_spike["Improvement"].astype("string").str.strip()

# ✅ 1) remove real missing values after explode
expanded_spike = expanded_spike.dropna(subset=["Encoding", "Improvement", key_col])

# ✅ 2) remove empty strings
expanded_spike = expanded_spike[
    (expanded_spike["Encoding"] != "") &
    (expanded_spike["Improvement"] != "") &
    (expanded_spike[key_col].astype("string").str.strip() != "")
]

# ✅ 3) remove literal placeholders
bad = {"<na>", "na", "nan", "none", "null"}
expanded_spike = expanded_spike[
    ~expanded_spike["Improvement"].str.lower().isin(bad)
]

# ✅ 4) IMPORTANT: prevent double counting
# Each paper contributes at most 1 to each (Encoding, Improvement)
expanded_spike = expanded_spike.drop_duplicates(subset=[key_col, "Encoding", "Improvement"])

# =========================
# Build table with Frequency + My_Key list
# =========================
table_spike = (
    expanded_spike
    .groupby(["Encoding", "Improvement"])
    .agg(
        Frequency=(key_col, "nunique"),
        My_Key_List=(key_col, lambda x: sorted(set(x)))
    )
    .reset_index()
)

table_spike = table_spike.sort_values(["Encoding", "Frequency"], ascending=[True, False])

# (Optional) convert list to a single string for LaTeX
table_spike["My_Key_List"] = table_spike["My_Key_List"].apply(lambda lst: ", ".join(lst))

# =========================
# Export to LaTeX
# =========================
os.makedirs("figs", exist_ok=True)

latex_spike = table_spike.to_latex(
    index=False,
    longtable=True,
    caption="Spike-wise efficiency improvements by encoding type (with supporting studies).",
    label="tab:encoding_spike_efficiency",
    escape=True
)

with open("figs/encoding_spike_efficiency.tex", "w", encoding="utf-8") as f:
    f.write(latex_spike)

print("figs/encoding_spike_efficiency.tex saved ✅")

table_spike
```

```python
import pandas as pd
import os

key_col = "My_Key"
enc_col = "RQ1a_Encoding_Type"
res_col = "RQ2b_Resource_efficiency"

# 1) keep rows that have Encoding and Resource text (initial filter)
data_res = df[[key_col, enc_col, res_col]].dropna(subset=[enc_col, res_col]).copy()

# 2) explode Encoding + Improvement by \n
expanded_res = (
    data_res
    .assign(
        Encoding=data_res[enc_col].astype("string").str.split("\n"),
        Improvement=data_res[res_col].astype("string").str.split("\n")
    )
    .explode("Encoding")
    .explode("Improvement")
)

# 3) strip
expanded_res["Encoding"] = expanded_res["Encoding"].astype("string").str.strip()
expanded_res["Improvement"] = expanded_res["Improvement"].astype("string").str.strip()
expanded_res[key_col] = expanded_res[key_col].astype("string").str.strip()

# 4) drop real missing values that can appear after explode
expanded_res = expanded_res.dropna(subset=[key_col, "Encoding", "Improvement"])

# 5) remove empty strings
expanded_res = expanded_res[
    (expanded_res[key_col] != "") &
    (expanded_res["Encoding"] != "") &
    (expanded_res["Improvement"] != "")
]

# 6) remove placeholder NA-like strings (in case they exist as text)
bad = {"<na>", "na", "nan", "none", "null"}
expanded_res = expanded_res[
    ~expanded_res["Improvement"].str.lower().isin(bad)
]

# 7) remove event-driven terms (since you decided to remove them)
remove_terms = {"event-driven computation", "event-driven processing"}
expanded_res = expanded_res[
    ~expanded_res["Improvement"].str.lower().isin(remove_terms)
]

# ✅ 8) prevent double counting:
# each paper contributes at most 1 to each (Encoding, Improvement)
expanded_res = expanded_res.drop_duplicates(subset=[key_col, "Encoding", "Improvement"])

# 9) build table with Frequency + My_Key list
table_resource = (
    expanded_res
    .groupby(["Encoding", "Improvement"])
    .agg(
        Frequency=(key_col, "nunique"),
        My_Key_List=(key_col, lambda x: sorted(set(x)))
    )
    .reset_index()
)

table_resource = table_resource.sort_values(
    ["Encoding", "Frequency"],
    ascending=[True, False]
)

# convert list to string for LaTeX
table_resource["My_Key_List"] = table_resource["My_Key_List"].apply(lambda lst: ", ".join(lst))

# 10) export to LaTeX
os.makedirs("figs", exist_ok=True)

latex_resource = table_resource.to_latex(
    index=False,
    longtable=True,
    caption="Resource efficiency improvements by encoding type (with supporting studies).",
    label="tab:encoding_resource_efficiency",
    escape=True
)

with open("figs/encoding_resource_efficiency.tex", "w", encoding="utf-8") as f:
    f.write(latex_resource)

print("figs/encoding_resource_efficiency.tex saved ✅")

table_resource
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# =========================
# Columns
# =========================
key_col   = "My_Key"
enc_col   = "RQ1a_Encoding_Type"
spike_col = "RQ2a_spike_wise_efficiency"

# =========================
# Controlled vocabulary (based on your provided column values)
# =========================
VOCAB_SPIKE = [
    "Sparse spike activity",
    "Reduced firing rate",
    "Reduced spike count",
    "Efficient Timesteps",
    "Activity-driven suppression",
    "Activity-driven responsiveness",
    "Stable spike behavior",
]

fig_dir = "figs"
os.makedirs(fig_dir, exist_ok=True)

BAD = {"<na>", "na", "nan", "none", "null", ""}

def clean_text(x):
    if pd.isna(x):
        return pd.NA
    s = str(x).strip()
    if s == "" or s.lower() in BAD:
        return pd.NA
    return s

def normalize_spike_token(t: str) -> str:
    """
    Normalize tokens to match YOUR controlled vocabulary exactly.
    Only standardizes simple variants (case/spacing).
    """
    if pd.isna(t):
        return ""
    s = str(t).strip()

    low = s.lower()

    # unify common variants
    if low == "sparse spike activity":
        return "Sparse spike activity"
    if low == "reduced firing rate":
        return "Reduced firing rate"
    if low == "reduced spike count":
        return "Reduced spike count"
    if low == "efficient timesteps":
        return "Efficient Timesteps"
    if low == "activity-driven suppression":
        return "Activity-driven suppression"
    if low == "activity-driven responsiveness":
        return "Activity-driven responsiveness"
    if low == "stable spike behavior":
        return "Stable spike behavior"

    # keep anything unexpected as-is (you can audit later)
    return s

# =========================
# 1) Prepare + explode encodings
# =========================
data = df[[key_col, enc_col, spike_col]].copy()
data[key_col]   = data[key_col].apply(clean_text)
data[enc_col]   = data[enc_col].apply(clean_text)
data[spike_col] = data[spike_col].apply(clean_text)

data = data.dropna(subset=[key_col, enc_col])  # need paper id + encoding at least

expanded = (
    data.assign(Encoding=data[enc_col].astype("string").str.split("\n"))
        .explode("Encoding")
)

expanded["Encoding"] = expanded["Encoding"].astype("string").str.strip()
expanded = expanded.dropna(subset=["Encoding"])
expanded = expanded[expanded["Encoding"] != ""]

# prevent inflated counting: one paper once per encoding
expanded = expanded.drop_duplicates(subset=[key_col, "Encoding"])

# =========================
# 2) Explode spike-wise efficiency tokens (multi-label)
# =========================
sp = expanded.dropna(subset=[spike_col]).copy()

sp = (
    sp.assign(sp_token=sp[spike_col].astype("string").str.split("\n"))
      .explode("sp_token")
)

sp["sp_token"] = sp["sp_token"].astype("string").str.strip()
sp = sp.dropna(subset=["sp_token"])
sp = sp[sp["sp_token"] != ""]

# Normalize to controlled vocabulary
sp["Spike_Eff_Token"] = sp["sp_token"].apply(normalize_spike_token)

# strict: keep only tokens in your vocab
sp = sp[sp["Spike_Eff_Token"].isin(VOCAB_SPIKE)]

# Count each paper once per (Encoding, Token)
sp = sp.drop_duplicates(subset=[key_col, "Encoding", "Spike_Eff_Token"])

# =========================
# 3.5) Mapping: (Encoding, Spike_Eff_Token) -> papers (My_Key)
# =========================

pair_to_papers = (
    sp.groupby(["Encoding", "Spike_Eff_Token"])[key_col]
      .apply(lambda s: sorted(set(s.dropna().astype(str))))
)

pair_map_df = pair_to_papers.reset_index(name="Papers_MyKey")
pair_map_df["Count_unique_papers"] = pair_map_df["Papers_MyKey"].apply(len)

pair_map_df = pair_map_df.sort_values(
    ["Count_unique_papers", "Encoding", "Spike_Eff_Token"],
    ascending=[False, True, True]
).reset_index(drop=True)

display(pair_map_df)

# Optional: save mapping table
out_map = os.path.join(fig_dir, "map_RQ2a_spike_efficiency_vs_RQ1a_encoding__papers.csv")
pair_map_df.to_csv(out_map, index=False)
print("Saved mapping ✅", out_map)

# Optional: dict form for quick lookup
pair_to_papers_dict = pair_to_papers.to_dict()


# =========================
# 3) Crosstab: Encoding × Token (exact terms)
# =========================
tab = pd.crosstab(sp["Encoding"], sp["Spike_Eff_Token"])

# ensure all vocab columns exist even if 0
for c in VOCAB_SPIKE:
    if c not in tab.columns:
        tab[c] = 0
tab = tab[VOCAB_SPIKE]

# sort encodings by total spike-wise reports
tab = tab.loc[tab.sum(axis=1).sort_values(ascending=False).index]

# =========================
# 4) Plot stacked bar
# =========================
ax = tab.plot(kind="bar", stacked=True, figsize=(12, 6))
ax.set_title("Spike-wise Efficiency (Controlled Vocabulary) vs Encoding Type (RQ2a × RQ1a)")
ax.set_xlabel("Encoding Type")
ax.set_ylabel("Number of studies")
ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
plt.xticks(rotation=30, ha="right")

ax.legend(title="RQ2a_spike_wise_efficiency", bbox_to_anchor=(1.02, 1),
          loc="upper left", borderaxespad=0.)

plt.tight_layout()
out_path = os.path.join(fig_dir, "fig_RQ2a_spike_efficiency_vocab_vs_encoding.png")
plt.savefig(out_path, dpi=300, bbox_inches="tight")
plt.show()

print("Saved ✅", out_path)
display(tab)

# =========================
# 5) OPTIONAL: audit unexpected tokens (helpful if anything gets excluded)
# =========================
# all_tokens = (
#     expanded[spike_col].dropna().astype(str).str.split("\n").explode().str.strip()
# )
# all_tokens = all_tokens[(all_tokens != "") & (~all_tokens.str.lower().isin(BAD))]
# normed = all_tokens.apply(normalize_spike_token)
# dropped = sorted(set(normed) - set(VOCAB_SPIKE))
# print("Tokens present but excluded (not in VOCAB_SPIKE):", dropped)
```
