```python
import pandas as pd
import os
from reportlab.lib.pagesizes import A4, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, PageBreak
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import cm

# -------------------------------
# Configuration
# -------------------------------
SOURCE_EXCEL_PATH = r"C:\Users\zohoo\OneDrive - Umm Al-Qura University\GP_Master_Degree\SLR_Data_Extraction_shorter_Last_included.xlsx"
SHEET_NAME = "data extraction & scores"

OUT_EXCEL = "Quality_Assessment_Table.xlsx"
OUT_CSV   = "Quality_Assessment_Table.csv"
OUT_PDF   = "Quality_Assessment_Table.pdf"

# Recommended important columns for QA reporting
IMPORTANT_COLS = [
    "My_Key","Title","DOI",
    "QA1. Purpose & Context", "QA2. Methodological Quality", "QA3. Data & Tools Transparency",
    "QA4. Rigor & Validity", "QA5. Results & Interpretation", "QA6. Contribution & Credibility",
    "Total QA Score"
]

# -------------------------------
# Load data
# -------------------------------
df = pd.read_excel(SOURCE_EXCEL_PATH, sheet_name=SHEET_NAME, engine="openpyxl")

# Keep only columns that actually exist (prevents crashes if headers differ slightly)
existing_cols = [c for c in IMPORTANT_COLS if c in df.columns]
missing_cols = [c for c in IMPORTANT_COLS if c not in df.columns]

qa_df = df[existing_cols].copy()

# Optional: sort for readability
if "My_Key" in qa_df.columns:
    qa_df = qa_df.sort_values(by=["My_Key"], kind="stable")

# -------------------------------
# Save Excel + CSV (latest only)
# -------------------------------
qa_df.to_excel(OUT_EXCEL, index=False)
qa_df.to_csv(OUT_CSV, index=False)

# -------------------------------
# Create a readable PDF table
# (Splits into chunks to avoid ultra-wide/too-long single tables)
# -------------------------------
styles = getSampleStyleSheet()
styleN = styles["Normal"]

def to_paragraph(x):
    """Convert cell content into a wrapped Paragraph for PDF rendering."""
    if pd.isna(x):
        return Paragraph("", styleN)
    text = str(x).replace("\n", "<br/>")
    return Paragraph(text, styleN)

# Convert dataframe to table data with wrapped cells
header = [to_paragraph(c) for c in qa_df.columns]
rows = [[to_paragraph(v) for v in row] for row in qa_df.values.tolist()]

# Page setup (landscape to fit more columns)
doc = SimpleDocTemplate(
    OUT_PDF,
    pagesize=landscape(A4),
    leftMargin=1.2*cm, rightMargin=1.2*cm,
    topMargin=1.2*cm, bottomMargin=1.2*cm
)

elements = []

# Split rows across multiple tables/pages if needed
CHUNK_SIZE = 18  # adjust if your rows are very long (titles/URLs)
for start in range(0, len(rows), CHUNK_SIZE):
    chunk = rows[start:start + CHUNK_SIZE]
    data = [header] + chunk

    # Auto-ish column widths (keeps PDF more stable)
    # You can fine-tune widths if needed.
    col_widths = [3.0*cm] * len(qa_df.columns)

    t = Table(data, colWidths=col_widths, repeatRows=1)

    t.setStyle(TableStyle([
        ("BACKGROUND", (0, 0), (-1, 0), colors.lightgrey),
        ("TEXTCOLOR", (0, 0), (-1, 0), colors.black),
        ("GRID", (0, 0), (-1, -1), 0.25, colors.grey),
        ("VALIGN", (0, 0), (-1, -1), "TOP"),
        ("FONTNAME", (0, 0), (-1, 0), "Helvetica-Bold"),
        ("FONTSIZE", (0, 0), (-1, -1), 8),
        ("ROWBACKGROUNDS", (0, 1), (-1, -1), [colors.whitesmoke, colors.white]),
        ("BOTTOMPADDING", (0, 0), (-1, 0), 6),
    ]))

    elements.append(t)
    if start + CHUNK_SIZE < len(rows):
        elements.append(PageBreak())

doc.build(elements)

print("QA exports created successfully:")
print(f"- Excel: {os.path.abspath(OUT_EXCEL)}")
print(f"- CSV:   {os.path.abspath(OUT_CSV)}")
print(f"- PDF:   {os.path.abspath(OUT_PDF)}")

if missing_cols:
    print("\nWARNING: The following requested columns were not found in the sheet:")
    for c in missing_cols:
        print(f" - {c}")
```

```python
import pandas as pd

def qa_level(score):
    if 0 <= score <= 4:
        return "Low"
    elif 5 <= score <= 8:
        return "Moderate"
    elif 9 <= score <= 12:
        return "High"
    else:
        return "Unknown"

df["QA_Level"] = df["Total QA Score"].apply(qa_level)

qa_table = (
    df
    .groupby(["QA_Level", "Total QA Score"])
    .agg(
        Frequency=("My_Key", "count"),
        My_Key=("My_Key", lambda x: ", ".join(sorted(x)))
    )
    .reset_index()
    .sort_values(
        by=["QA_Level", "Total QA Score"],
        ascending=[False, False]
    )
)

latex_table = qa_table.to_latex(
    index=False,
    escape=False,          # مهم لعدم كسر My_Key
    column_format="llrp{8cm}",
    caption="Distribution of studies by total quality assessment (QA) score.",
    label="tab:qa_distribution"
)

fig_dir = "figs"

with open(fig_dir+"\qa_distribution.tex", "w", encoding="utf-8") as f:
    f.write(latex_table)
print(fig_dir+"\qa_distribution.tex ✅")

qa_table

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

def safe_filename(name):
    name = re.sub(r"[^\w\-. ]+", "_", name)
    return name.replace(" ", "_")

def analyze_tokens_with_keys_pie(
    df,
    col,
    key_col="My_Key",
    fig_dir="figs",
    unique_per_paper=True,
    filename=None,
    title=None,
    min_pct=2      # hide very small slices (optional)
):
    os.makedirs(fig_dir, exist_ok=True)

    tmp = df[[key_col, col]].dropna(subset=[col]).copy()
    tmp[col] = tmp[col].astype(str)

    # split ONLY by newline
    tokens_df = (
        tmp
        .assign(Token=tmp[col].str.split("\n"))
        .explode("Token")
    )

    tokens_df["Token"] = tokens_df["Token"].str.strip()
    tokens_df = tokens_df[tokens_df["Token"] != ""]

    # remove duplicates within same paper
    if unique_per_paper:
        tokens_df = tokens_df.drop_duplicates(subset=[key_col, "Token"])

    # Token → [My_Key]
    mapping = (
        tokens_df
        .groupby("Token")[key_col]
        .apply(list)
        .sort_index()
    )

    # Frequency
    freq = tokens_df["Token"].value_counts()

    if filename is None:
        filename = safe_filename(col) + "_pie.png"
    if title is None:
        title = f"Distribution of {col}"

    # ---- PIE PLOT ----
    plt.figure(figsize=(9, 9))

    def autopct(pct):
        total = freq.sum()
        count = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n(n={count})" if pct >= min_pct else ""

    freq.plot(
        kind="pie",
        autopct=autopct,
        startangle=90,
        counterclock=False
    )

    plt.title(title)
    plt.ylabel("")  # remove default ylabel
    plt.tight_layout()

    outpath = os.path.join(fig_dir, filename)
    plt.savefig(outpath, dpi=300, bbox_inches="tight")
    plt.show()

    print("==== Token → [My_Key] ====")
    print(mapping)
    print("\n==== Frequency ====")
    print(freq)
    print(f"\n{filename} saved ✅")

    return tokens_df, mapping, freq

results_pie = {}

cols_to_run = [
    # "RQ1a_Encoding_Type",
    # "RQ1b_Learnable_encoding_parameters",
    # "RQ1c_Training_mechanism",
    "RQ1d_Hybrid_Integration_Type",
    # "RQ2a_spike_wise_efficiency",
    # "RQ2b_Resource_efficiency",
    # "RQ2c_Training_Behavior",
    # "RQ2d_Model_Behavior",
    "RQ3a_Dataset_modality",
    # "RQ3b_Input_Representation",
    # "RQ3c_Datasets_Used",
    # "RQ3d_Task_or_Domain",
    # "RQ3e_Evaluation_Metrics",
    # "RQ4a_Limitation_Category",
    # "RQ4b_Reported_Limitations",
    # "RQ4c_Research_Gaps",
    # "hardware_or_simulator",
    "paper_type",
    # Architecture_Type",
    # "code_availability",
    # "training_config_short",
]

for c in cols_to_run:
    tokens_df, mapping, freq = analyze_tokens_with_keys_pie(
        df,
        col=c,
        fig_dir=fig_dir,
        filename=f"{safe_filename(c)}_pie.png",
        title=f"Distribution of {c}"
    )
    results_pie[c] = {
        "mapping": mapping,
        "freq": freq
    }
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# =========================
# Columns
# =========================
key_col   = "My_Key"
enc_col   = "RQ1a_Encoding_Type"
spike_col = "RQ2a_spike_wise_efficiency"
res_col   = "RQ2b_Resource_efficiency"

# folder
fig_dir = "figs"
os.makedirs(fig_dir, exist_ok=True)

# =========================
# Load + Expand (paper-level)
# =========================
data = df[[key_col, enc_col, spike_col, res_col]].dropna(subset=[enc_col]).copy()

expanded = (
    data
    .assign(Encoding=data[enc_col].astype("string").str.split("\n"))
    .explode("Encoding")
)

expanded["Encoding"] = expanded["Encoding"].astype("string").str.strip()
expanded[key_col] = expanded[key_col].astype("string").str.strip()

# drop empty keys/encodings
expanded = expanded.dropna(subset=[key_col, "Encoding"])
expanded = expanded[(expanded[key_col] != "") & (expanded["Encoding"] != "")]

# =========================
# Clean spike/resource text -> true NA
# =========================
bad = {
    "<na>", "na", "nan", "none", "null",
    "n/a", "not applicable", "not reported",
    "no", "no improvement", "none reported"
}

for c in [spike_col, res_col]:
    expanded[c] = expanded[c].astype("string").str.strip()
    expanded.loc[expanded[c] == "", c] = pd.NA
    expanded.loc[expanded[c].str.lower().isin(bad), c] = pd.NA

# ✅ prevent inflated counting:
# one paper counted once per encoding
expanded = expanded.drop_duplicates(subset=[key_col, "Encoding"])

# =========================
# Flags (reported? yes/no)
# =========================
expanded["Spike_Improvement"] = expanded[spike_col].notna().astype(int)
expanded["Resource_Improvement"] = expanded[res_col].notna().astype(int)
expanded["Both_Improvements"] = (
    (expanded["Spike_Improvement"] == 1) &
    (expanded["Resource_Improvement"] == 1)
).astype(int)
expanded["Spike_Only"] = (
    (expanded["Spike_Improvement"] == 1) &
    (expanded["Resource_Improvement"] == 0)
).astype(int)
expanded["Resource_Only"] = (
    (expanded["Spike_Improvement"] == 0) &
    (expanded["Resource_Improvement"] == 1)
).astype(int)

# =========================
# Pivots (counts of papers)
# =========================
pivot_spike = pd.pivot_table(
    expanded, index="Encoding", values="Spike_Improvement", aggfunc="sum"
)
pivot_resource = pd.pivot_table(
    expanded, index="Encoding", values="Resource_Improvement", aggfunc="sum"
)

# common order
order = (
    pivot_spike.join(pivot_resource, how="outer")
    .fillna(0)
    .sum(axis=1)
    .sort_values(ascending=False)
    .index
)
pivot_spike = pivot_spike.loc[order]
pivot_resource = pivot_resource.loc[order]

# =========================
# My_Key mappings
# =========================
spike_keys = (
    expanded.loc[expanded["Spike_Improvement"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
resource_keys = (
    expanded.loc[expanded["Resource_Improvement"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
both_keys = (
    expanded.loc[expanded["Both_Improvements"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
spike_only_keys = (
    expanded.loc[expanded["Spike_Only"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)
resource_only_keys = (
    expanded.loc[expanded["Resource_Only"] == 1]
    .groupby("Encoding")[key_col].apply(list).sort_index()
)

# =========================
# Sanity checks (optional but useful)
# =========================
print("Spike=1 count:", int(expanded["Spike_Improvement"].sum()))
print("Resource=1 count:", int(expanded["Resource_Improvement"].sum()))
print("Both=1 count:", int(expanded["Both_Improvements"].sum()))
print("Spike-only=1 count:", int(expanded["Spike_Only"].sum()))
print("Resource-only=1 count:", int(expanded["Resource_Only"].sum()))

# =========================
# Plot side-by-side
# =========================
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

pivot_spike.plot(kind="bar", ax=axes[0], legend=False)
axes[0].set_title("Spike-wise Efficiency Improvements (count of papers)")
axes[0].set_xlabel("Encoding Type")
axes[0].set_ylabel("Number of Papers")
axes[0].tick_params(axis="x", rotation=45)
axes[0].yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
axes[0].yaxis.grid(True, linestyle="--", alpha=0.6)

pivot_resource.plot(kind="bar", ax=axes[1], legend=False)
axes[1].set_title("Resource Efficiency Improvements (count of papers)")
axes[1].set_xlabel("Encoding Type")
axes[1].tick_params(axis="x", rotation=45)
axes[1].yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
axes[1].yaxis.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()

out_path = os.path.join(fig_dir, "Resource_Spike_wise_Efficiency_Improvements.png")
plt.savefig(out_path, dpi=300, bbox_inches="tight")
print(f"{out_path} saved ✅")
plt.show()

# =========================
# Print My_Key for discussion
# =========================
print("\n==== Encoding → My_Key (Spike-wise improvements) ====")
print(spike_keys)

print("\n==== Encoding → My_Key (Resource improvements) ====")
print(resource_keys)

print("\n==== Encoding → My_Key (Both improvements) ====")
print(both_keys)

print("\n==== Encoding → My_Key (Spike-only) ====")
print(spike_only_keys)

print("\n==== Encoding → My_Key (Resource-only) ====")
print(resource_only_keys)
```
